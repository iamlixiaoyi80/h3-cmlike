# 🎮 H3-CMlike Android APK 云端构建

这个 Notebook 会自动：
1. 安装 Node.js 和 Android SDK
2. 克隆 H3-CMlike 项目
3. 配置 Capacitor
4. 构建 Android APK
5. 提供下载链接

**使用方法：** 依次运行每个单元格即可

## 1️⃣ 安装 Node.js 20

```python
# 安装 Node.js 20
!curl -fsSL https://deb.nodesource.com/setup_20.x | sudo -E bash -
!sudo apt-get install -y nodejs

# 验证安装
!node --version
!npm --version
```

## 2️⃣ 安装 Android SDK 命令行工具

```python
# 创建 Android SDK 目录
!mkdir -p ~/android-sdk/cmdline-tools

# 下载并解压命令行工具
!cd ~/android-sdk/cmdline-tools && \
  wget -q https://dl.google.com/android/repository/commandlinetools-linux-11076708_latest.zip -O tools.zip && \
  unzip -q tools.zip && \
  mv cmdline-tools latest && \
  rm tools.zip

# 设置环境变量
import os
os.environ['ANDROID_SDK_ROOT'] = os.path.expanduser('~/android-sdk')
os.environ['PATH'] = os.environ['ANDROID_SDK_ROOT'] + '/cmdline-tools/latest/bin:' + os.environ['PATH']

# 接受许可证并安装必要组件
!yes | sdkmanager --licenses > /dev/null 2>&1
!sdkmanager "platform-tools" "platforms;android-34" "build-tools;34.0.0" > /dev/null 2>&1

print("✅ Android SDK 安装完成！")
```

## 3️⃣ 克隆项目并安装依赖

```python
# 克隆项目
%cd /content
!rm -rf h3-cmlike
!git clone https://github.com/iamlixiaoyi80/h3-cmlike.git
%cd h3-cmlike

# 安装依赖
!npm install

print("✅ 项目克隆和依赖安装完成！")
```

## 4️⃣ 安装 Capacitor 并初始化

```python
# 安装 Capacitor
!npm install @capacitor/core @capacitor/cli @capacitor/android

# 初始化 Capacitor（如果还没有配置）
!npx cap init "H3-CMlike" "com.h3cmlike.game" --web-dir dist

print("✅ Capacitor 初始化完成！")
```

## 5️⃣ 构建 Vue 项目并同步

```python
# 构建 Vue 项目
!npm run build

# 添加 Android 平台（如果还没有）
!npx cap add android

# 同步到 Android
!npx cap sync android

print("✅ 构建和同步完成！")
```

## 6️⃣ 构建 APK

```python
# 设置 Android SDK 路径
import os
os.environ['ANDROID_SDK_ROOT'] = os.path.expanduser('~/android-sdk')

# 创建 local.properties 文件
with open('android/local.properties', 'w') as f:
    f.write(f"sdk.dir={os.environ['ANDROID_SDK_ROOT']}")

# 构建 Debug APK
%cd android
!chmod +x gradlew
!./gradlew assembleDebug --no-daemon

print("✅ APK 构建完成！")
```

## 7️⃣ 下载 APK

```python
import shutil
from google.colab import files

# 复制 APK 到根目录
apk_path = 'app/build/outputs/apk/debug/app-debug.apk'
output_path = '/content/h3-cmlike-debug.apk'
shutil.copy(apk_path, output_path)

# 显示文件信息
!ls -lh {output_path}

# 下载
files.download(output_path)

print("🎉 APK 已准备好，浏览器会自动下载！")
```
